```python
import time
from memory_profiler import memory_usage


def benchmark(func, *args, **kwargs):
    # Measure memory usage before running the function
    mem_before = memory_usage()[0]

    # Measure time before running the function
    start_time = time.time()

    # Run the function
    result = func(*args, **kwargs)

    # Measure time after running the function
    end_time = time.time()

    # Measure memory usage after running the function
    mem_after = memory_usage()[0]

    # Calculate time and memory usage
    time_taken = end_time - start_time
    mem_used = mem_after - mem_before

    print(f"Time taken: {time_taken:.6f} seconds")
    print(f"Memory used: {mem_used:.6f} MiB")

    return result


# Example usage
def example_function(n):
    return [i**2 for i in range(n)]


# Benchmark the example_function
benchmark(example_function, 1000000)
```

#14 | Longest Collatz Sequence


```python
cache = dict()


def rec(n):
    if n == 1:
        return 1

    if n in cache:
        return cache[n]
    if n % 2 == 0:
        out = 1 + rec(n / 2)
    else:
        out = 1 + rec(3 * n + 1)
    cache[n] = out
    return out


def longCollatz():
    maxi = 1
    digit = 1
    for i in range(1, 1000001):
        if rec(i) > maxi:
            maxi = rec(i)
            digit = i

    return digit


# print(longCollatz())
benchmark(longCollatz)
```

#15 | Digit sum


```python
def digitSum(n):
    sum = 0
    values = [0] * n
    values[0] = 1
    carry = [0] * n
    while True:
        isEmpty = True
        for i in range(1, n):
            val = values[i - 1] * 2 + carry[i]
            if val != 0:
                isEmpty = False
            carry[i] = (int)((val - val % 10) / 10)
            values[i] = val % 10
        # print("values->", values)
        # print("carry->", carry)
        sum += values[n - 1]
        if isEmpty:
            break
        values = [0] * n
    return sum


print(digitSum(1001))
```

#20 | Factorial Digit Sum


```python
def digitSum(n):
    sum = 0
    values = [0] * n
    values[0] = 1
    carry = [0] * n

    while True:
        isEmpty = True
        for i in range(1, n):
            val = values[i - 1] * (i + 1) + carry[i]
            if val != 0:
                isEmpty = False
            carry[i] = (int)((val - val % 10) / 10)
            values[i] = val % 10
        # print("values->", values)
        # print("carry->", carry)
        sum += values[n - 1]
        if isEmpty:
            break
        values = [0] * n
    return sum


benchmark(digitSum, 100)
```

```python
import math


def get_divisors_sum(n):
    sum = 1
    i = 2
    while i * i < n:
        if n % i == 0:
            sum += i
            sum += n / i
        i += 1

    return sum


def amicable(n):
    mapping = dict()
    store = [0]
    sum = 0
    for i in range(1, n):
        div = int(get_divisors_sum(i))
        store.append(div)
        if div in mapping:
            mapping[div].append(i)
        else:
            mapping[div] = [i]
    # iii
    for i in range(2, n):
        div = store[i]

        if i not in mapping:
            continue

        for t in mapping[i]:
            if t == div and t != i:
                sum += t
    return sum


benchmark(amicable, 10000)
```

```python
import string


def get_next_quotient(di, dd):
    r = dd % di
    return ((dd - r) / di, r)


def getMaxRepeating(number):
    remainders = dict()
    stri = ""
    di = number
    dd = 1
    count = 0
    while True:
        (q, r) = get_next_quotient(di, dd)
        dd = r * 10
        count += 1
        stri += str(int(q))
        if r == 0:
            return 0
        if remainders.get(r):
            return count - remainders.get(r)
        else:
            remainders[r] = count


# print(int("9"))
# print(1/23)
# print(getMaxRepeating(983))


def getMaxRepeatingInRange(number):
    maxi = 0
    maxNum = 2
    for i in range(2, number):
        cur = getMaxRepeating(i)
        if maxi < cur:
            maxi = cur
            maxNum = i

    return (maxi, maxNum)


benchmark(getMaxRepeatingInRange, 1000)
```

#29 Distinct Powers


```python
def distinctPowers(n):
    sum = 0
    checkList = [False] * (n + 1)
    i = 2
    while i <= n:
        if checkList[i]:
            i += 1
            continue
        exp_set = set()
        base = i
        base_exp = 1
        while base <= n:
            checkList[base] = True
            for t in range(2, n + 1):
                exp_set.add(base_exp * t)
            base_exp += 1
            base *= i
        sum += len(exp_set)
        i += 1

    print(sum)


benchmark(distinctPowers, 100)
```

#31 Coin Sums


```python
coins = [200, 100, 50, 20, 10, 5, 2, 1]


def coin_sum(num, index):
    if num < 0:
        return 0
    if num == 0:
        return 1
    if index >= len(coins):
        return False

    count = 0

    while num >= 0:
        count += coin_sum(num, index + 1)
        num -= coins[index]

    return count


coin_sum(200, 0)
```
